### Creating price category

```python
# finding value counts of price column
df['price'].value_counts (dropna= False)
```

```python
df.describe()
```

```python
# Defining function to create price category

def price_label(row):
    if row['price'] <=86:
        return 'Low-range product'
    elif (row['price']>86) and (row['price']<=250):
        return 'Mid-range product'
    elif row['price'] >250:
        return 'High range'
    else: return 'Not listed'
```

```python
#creating price range column based off function
df['price_range']=df.apply(price_label, axis=1)
```

```python
df.head()
```

```python
#creating catplot of prices

g = sns.catplot(x='number_of_reviews', y='price', hue='price_range', data=df)
```

```python
#plotting price catplot without 'high-range' outliers 

g = sns.catplot(x='number_of_reviews', y='price', hue='price_range', data=df)

plt.ylim(0, 13000)
plt.show()
```
