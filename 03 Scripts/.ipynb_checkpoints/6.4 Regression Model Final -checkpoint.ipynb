# Step 1: Install the necessary libraries


# Step 1 & 2

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
# Load your data
df= pd.read_pickle(r'/Users/evancarr/Desktop/New York Python Project/02 Prepared Data/6.2 Exploring Relationships.pkl')
```

# 3. Cleaning Data


```python
# Drop unnecessary columns
df = df.dropna(subset=['price', 'minimum_nights', 'reviews_per_month', 'latitude', 'longitude', 'number_of_reviews', 'room_type', 'last_review', 'license'])

# Drop rows with missing values in 'price'
df['log_price'] = np.log(df['price'])
```

```python
# Filter extreme values for 'price'
# Filter extreme values for 'log_price' instead of 'price'
upper_limit_log_price = df['log_price'].quantile(0.99)
df = df[df['log_price'] <= upper_limit_log_price]

# Check if any missing values remain
print(df.isnull().sum())
```

# 4. Explore Visual Data

```python
sns.scatterplot(data=df, x='minimum_nights', y='price')
plt.title('Price vs Minimum Nights')
plt.xlabel('Minimum Nights')
plt.ylabel('Price')
plt.show()
```

```python
sns.boxplot(data=df, x='room_type', y='price')
plt.title('Price by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.show()
```

# 5. Hypothesis
### Hypothesis: "The price of a listing is influenced by the number of minimum nights, the room type, and the neighborhood."

# 6. Reshape variables for linear regression

```python
# Encode categorical variables 
df = pd.get_dummies(df, columns=['room_type', 'neighbourhood'], drop_first=True)
```

```python
# Create an interaction term 
df['min_nights_room_type'] = df['minimum_nights'] * df['room_type_Shared room']  

# Define independent variables (X) and dependent variable (y)
X = df[['minimum_nights', 'latitude', 'longitude', 'number_of_reviews', 'reviews_per_month', 
        'min_nights_room_type', 'last_review', 'license'] + [col for col in df.columns if 'room_type_' in col or 'neighbourhood_' in col]]

# Define the dependent variable (log_price)
df['log_price'] = np.log(df['price'])

y = df['log_price']

# Check the shape of X and y to ensure it's correct
print(X.shape, y.shape)
```

```python
X = pd.get_dummies(X, drop_first=True)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# 7. Split the data into training and test sets

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

# 8. Run linear regression

```python
# Initialize the Linear Regression model
model = LinearRegression()
```

```python
# Fit the model to the training data
model.fit(X_train, y_train)
```

# 9. Create plot to show the regression line on the test set

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

# 10. Interpret the regression line
### In the plot above, the regression line is the diagonal line from bottom-left to top-right.
### The closer the predicted values (points) are to this line, the better the model's predictions.

```python
y_pred_price = np.exp(y_pred)  # Transform back to original price scale
y_test_price = np.exp(y_test)  # Transform true values back to original scale
```

# 11. Check model performance statistics - MSE and R2 score

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error (MSE): {mse}')
print(f'R-squared: {r2}')
```

```python
# Visualize the predictions vs actual values
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.6)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linewidth=2)
plt.title('Predicted vs Actual Prices (Log-transformed)', fontsize=14)
plt.xlabel('Actual Prices (Log-transformed)')
plt.ylabel('Predicted Prices (Log-transformed)')
plt.show()
```

```python
# Cross-validation to check model robustness
cv_scores = cross_val_score(model, X_scaled, y, cv=5, scoring='neg_mean_squared_error')
print(f"Cross-validation MSE: {np.mean(-cv_scores)}")
```

# 12. Compare predicted y values with the actual y values in a DataFrame

```python
results = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
results['Actual'] = np.exp(results['Actual'])  # Transform back to original price scale
results['Predicted'] = np.exp(results['Predicted'])  # Transform back to original price scale

print(results.head())
```
